## Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
#x = range(-300, 300)
x = range(0,50)
x2 = range(-40,40)
y1=[math.sqrt(num) for num in x]
y2=[(num * num * num) for num in x2]
y3=[(2 ** num) for num in x]
y4=[(1/(num+1)) for num in x]

plt.figure(figsize = (8,8))

plt.subplot(221)
plt.plot(y1)
plt.title('Square root of X')

plt.subplot(222)
plt.plot(y2)
plt.title('X^3')

plt.subplot(223)
plt.plot(y3)
plt.title('2^X')

plt.subplot(224)
plt.plot(y4)
plt.title('1/(x + 1)')
```

Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# %matplotlib notebook
plt.figure(figsize = (8,8))

plt.subplot(221)
plt.plot(y1, color = 'r')
plt.legend("$")
plt.title('Square root of X')

plt.subplot(222)
plt.plot(y2, color = 'b')
plt.legend("$")
plt.title('X^3')

plt.subplot(223)
plt.plot(y3, color = 'g')
plt.legend("$")
plt.title('2^X')

plt.subplot(224)
plt.plot(y4, color = 'y')
plt.legend("$")
plt.title('1/(x + 1)')
```
